```python
#Creating line charts depicting proportion of male and female graduates from STEM fields

%matplotlib inline

#importing packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#reading in database as dataframe
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18,3))

#looping through six times
for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], linewidth=3)
    ax.set(xlim=(1968, 2011), ylim=(0,100), title=stem_cats[sp])
    for key,spine in ax.spines.items(): #setting spines to be invisible
        spine.set_visible(False)
    ax.tick_params(bottom='off', top='off', left='off', right='off') #not showing ticks
    #labeling lines
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
#creating line charts for all 17 degrees

fig = plt.figure(figsize=(15,25))

#making a list of degree names for different degree categories
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 
                 'Social Sciences and History', '']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
all_cats = []

#making a list of all majors in order to make the for loop process more concise
for i in range(6):
    all_cats.append(stem_cats[i])
    all_cats.append(lib_arts_cats[i])
    all_cats.append(other_cats[i])

#looping through list
for i, major in enumerate(all_cats):
    if major != '': #if a degree is in the list
        ax = fig.add_subplot(6,3,i+1)
        ax.plot(women_degrees['Year'], women_degrees[major], linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[major], linewidth=3)
        ax.set_title(major, fontweight="bold")
        ax.set(xlim=(1968,2011), ylim=(0,100), yticks=[0,100])
        #hiding spines
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        #hiding ticks
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3) #drawing a mid-horizontal line
        
        #putting in labels for Men and Women
        if i == 0:
            ax.text(2003, 82, 'Women')
            ax.text(2005, 15, 'Men')
        elif i == 1:
            ax.text(2003, 75, 'Women')
            ax.text(2005, 22, 'Men')
        elif i == 2:
            ax.text(2003, 90, 'Women')
            ax.text(2005, 8, 'Men')
        elif i ==15:
            ax.text(2005, 88, 'Men')
            ax.text(2003, 10, 'Women')
        elif i == 17:
            ax.text(2005, 62, 'Men')
            ax.text(2003, 36, 'Women')
    
plt.savefig('gender_distribution_all_degrees.png') #saving plots as png
plt.show()
```

Degrees that have not had much change in the proportion of men and women are Foreign Languages, Art and Performance, Education, English, and Health Professionals. Agriculture, Business, Social Siences and History and Architecture have narrowed the gap between two genders. The degrees that have gone through a transformation in the proportion of the genders are Communications and Journalism, Biology and Psychology; they were all predominantly male in the past and are now predominantly female. Computer Science and Engineering continues to be predominantly male, whereas Psychology, Education, Health Professions, English, Public Administration, Communications and Journalism, and Art and Performance are mostly female.
